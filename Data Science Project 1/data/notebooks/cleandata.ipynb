```python
import pandas as pd


url = "https://kworb.net/ww/"


tables = pd.read_html(url)


print(f"Total tables found: {len(tables)}")


df = tables[0]


df.head()
```

```python
print(df.columns)
df.head(10)
```

We extract the artist and title data to split artist and title

```python
df[['Artist', 'Title']] = df['Artist and Title'].str.split(' - ', n=1, expand=True)

df[['Artist', 'Title']]
```

Remove whitespace from the strings in these columns

```python
df['Artist'] = df['Artist'].str.strip()
df['Title'] = df['Title'].str.strip()

df[['Artist', 'Title']]
```

We're all good!

```python
print(df['Title'].isna().sum())
```

We create a column where all text is lowercase for matching purposes (the logic of the project starts here)

```python
df['Artist_clean'] = df['Artist'].str.lower().str.strip()
df['Title_clean'] = df['Title'].str.lower().str.strip()

df
```

```python
df[df['Artist_clean'].str.contains(r'\bft\.?\b|\bfeat\.?\b|\bfeaturing\b|&', regex=True, na=False)]
```

```python
df['Artist_clean'] = df['Title_clean'].str.replace(r'\bft\.?\b', 'feat', regex=True)
df['Artist_clean'] = df['Title_clean'].str.replace(r'\bfeaturing\b', 'feat', regex=True)

df
```
