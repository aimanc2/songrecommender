We should learn about how sklearn works conceptually tomorrow

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
# Standard Scaler is used to scale the features to have a mean of 0 and a standard deviation of 1. This is important because many machine learning algorithms perform better when the features are on a similar scale.

scaled_features = scaler.fit_transform(final_df[feature_cols])
# Data is analysed to learn that the mean and standard deviation of each feature, and then the data is transformed to have a mean of 0 and a standard deviation of 1.
```

```python
# Turn it back into a dataframe with the same column names and index as the original dataframe
# Feature_cols come from the original dataframe, and the index is preserved to maintain the relationship between the scaled features and the original data.

scaled_df = pd.DataFrame(scaled_features, columns=feature_cols, index=final_df.index)
scaled_df.head()
```
